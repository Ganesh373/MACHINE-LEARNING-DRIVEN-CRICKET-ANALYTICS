```python
chase_df = chase_df.sort_values(by=['match_id', 'over', 'ball']).reset_index(drop=True)
```

```python
chase_df['current_score'] = chase_df.groupby('match_id')['total_runs'].cumsum()
chase_df[['match_id', 'over', 'ball', 'total_runs', 'current_score']].head(15)
```

```python
chase_df['runs_required'] = chase_df['target_runs'] - chase_df['current_score']
chase_df['runs_required'] = chase_df['runs_required'].clip(lower=0)

chase_df[['match_id', 'over', 'ball', 'target_runs', 'current_score', 'runs_required']].head(15)
```

```python
chase_df['balls_bowled'] = (chase_df['over'] - 1) * 6 + chase_df['ball']
chase_df['balls_remaining'] = 120 - chase_df['balls_bowled']
chase_df['balls_remaining'] = chase_df['balls_remaining'].clip(lower=0)

chase_df[['match_id', 'over', 'ball', 'balls_bowled', 'balls_remaining']].head(15)
```

```python
chase_df['overs_remaining'] = chase_df['balls_remaining'] / 6
chase_df[['match_id', 'over', 'ball', 'balls_remaining', 'overs_remaining']].head(15)
```

```python
chase_df['is_wicket'] = chase_df['is_wicket'].fillna(0)

chase_df['wickets_fallen'] = chase_df.groupby('match_id')['is_wicket'].cumsum()
chase_df['wickets_left'] = 10 - chase_df['wickets_fallen']
chase_df['wickets_left'] = chase_df['wickets_left'].clip(lower=0)

chase_df[['match_id', 'over', 'ball', 'is_wicket', 'wickets_fallen', 'wickets_left']].head(20)
```

```python
chase_df['overs_bowled'] = chase_df['balls_bowled'] / 6

chase_df['current_run_rate'] = np.where(
    chase_df['overs_bowled'] > 0,
    chase_df['current_score'] / chase_df['overs_bowled'],
    0
)

chase_df['required_run_rate'] = np.where(
    chase_df['overs_remaining'] > 0,
    chase_df['runs_required'] / chase_df['overs_remaining'],
    0
)

chase_df[['match_id', 'over', 'ball', 'current_score', 'current_run_rate', 'runs_required', 'required_run_rate']].head(15)
```

```python
chase_df['is_dot'] = (chase_df['total_runs'] == 0).astype(int)
chase_df['is_boundary'] = ((chase_df['batsman_runs'] == 4) | (chase_df['batsman_runs'] == 6)).astype(int)

chase_df[['match_id', 'over', 'ball', 'total_runs', 'batsman_runs', 'is_dot', 'is_boundary']].head(15)
```

```python
chase_df['batter_ball_number'] = chase_df.groupby(['match_id', 'batter']).cumcount() + 1

chase_df[['match_id', 'batter', 'over', 'ball', 'batter_ball_number']].head(20)
```

```python
chase_df['new_batter_at_crease'] = (chase_df['batter_ball_number'] <= 5).astype(int)
chase_df['set_batter'] = (chase_df['batter_ball_number'] >= 15).astype(int)

chase_df[['match_id', 'batter', 'over', 'ball', 'batter_ball_number', 'new_batter_at_crease', 'set_batter']].head(20)
```

```python
def get_phase(over):
    if over <= 6:
        return 'Powerplay'
    elif over <= 15:
        return 'Middle'
    else:
        return 'Death'

chase_df['phase'] = chase_df['over'].apply(get_phase)
chase_df[['match_id', 'over', 'ball', 'phase']].head(20)
```

```python
chase_df['recent_runs_last_6'] = (
    chase_df.groupby('match_id')['total_runs']
    .transform(lambda x: x.rolling(window=6, min_periods=1).sum())
)

chase_df[['match_id', 'over', 'ball', 'total_runs', 'recent_runs_last_6']].head(20)
```

```python
chase_df['recent_wickets_last_12'] = (
    chase_df.groupby('match_id')['is_wicket']
    .transform(lambda x: x.rolling(window=12, min_periods=1).sum())
)

chase_df[['match_id', 'over', 'ball', 'is_wicket', 'recent_wickets_last_12']].head(20)
```

```python
chase_df['recent_dots_last_6'] = (
    chase_df.groupby('match_id')['is_dot']
    .transform(lambda x: x.rolling(window=6, min_periods=1).sum())
)

chase_df[['match_id', 'over', 'ball', 'is_dot', 'recent_dots_last_6']].head(20)
```

```python
chase_df['recent_boundaries_last_6'] = (
    chase_df.groupby('match_id')['is_boundary']
    .transform(lambda x: x.rolling(window=6, min_periods=1).sum())
)

chase_df[['match_id', 'over', 'ball', 'is_boundary', 'recent_boundaries_last_6']].head(20)
```

```python
bowler_over_runs = (
    chase_df.groupby(['match_id', 'bowler', 'over'], as_index=False)['total_runs']
    .sum()
    .rename(columns={'total_runs': 'runs_in_over'})
)

bowler_over_runs['prev_over_expensive_runs'] = (
    bowler_over_runs.groupby(['match_id', 'bowler'])['runs_in_over'].shift(1)
)

bowler_over_runs['prev_over_expensive_runs'] = bowler_over_runs['prev_over_expensive_runs'].fillna(0)

chase_df = chase_df.merge(
    bowler_over_runs[['match_id', 'bowler', 'over', 'prev_over_expensive_runs']],
    on=['match_id', 'bowler', 'over'],
    how='left'
)

chase_df[['match_id', 'bowler', 'over', 'prev_over_expensive_runs']].head(20)
```

```python
chase_df['wicket_block'] = chase_df.groupby('match_id')['is_wicket'].cumsum()

chase_df['partnership_runs'] = (
    chase_df.groupby(['match_id', 'wicket_block'])['total_runs'].cumsum()
)

chase_df[['match_id', 'over', 'ball', 'is_wicket', 'wicket_block', 'partnership_runs']].head(25)
```

```python
chase_df['chase_success'] = (chase_df['batting_team'] == chase_df['winner']).astype(int)

chase_df[['match_id', 'batting_team', 'winner', 'chase_success']].drop_duplicates().head(10)
```

```python
chase_df['comfortable_chase'] = (
    (chase_df['required_run_rate'] <= 7) &
    (chase_df['wickets_left'] >= 6)
).astype(int)

chase_df[['match_id', 'over', 'ball', 'required_run_rate', 'wickets_left', 'comfortable_chase']].head(20)
```

```python
chase_df['high_pressure_chase'] = (
    (chase_df['required_run_rate'] > 10) &
    (chase_df['balls_remaining'] <= 30)
).astype(int)

chase_df[['match_id', 'over', 'ball', 'required_run_rate', 'balls_remaining', 'high_pressure_chase']].head(20)
```

```python
chase_df['death_finish'] = (
    (chase_df['over'] >= 16) &
    (chase_df['runs_required'] > 0)
).astype(int)

chase_df[['match_id', 'over', 'ball', 'runs_required', 'death_finish']].head(20)
```

```python
chase_df['collapse_situation'] = (
    (chase_df['recent_wickets_last_12'] >= 2) |
    (chase_df['wickets_left'] <= 4)
).astype(int)

chase_df[['match_id', 'over', 'ball', 'recent_wickets_last_12', 'wickets_left', 'collapse_situation']].head(20)
```

```python
chase_df['controlled_chase'] = (
    (chase_df['required_run_rate'] > 7) &
    (chase_df['required_run_rate'] <= 10) &
    (chase_df['wickets_left'] >= 5)
).astype(int)

chase_df[['match_id', 'over', 'ball', 'required_run_rate', 'wickets_left', 'controlled_chase']].head(20)
```

```python
def classify_match_situation(row):
    if row['collapse_situation'] == 1:
        return 'Collapse Situation'
    elif row['high_pressure_chase'] == 1:
        return 'High Pressure Chase'
    elif row['death_finish'] == 1:
        return 'Death Finish'
    elif row['comfortable_chase'] == 1:
        return 'Comfortable Chase'
    elif row['controlled_chase'] == 1:
        return 'Controlled Chase'
    else:
        return 'Neutral Situation'

chase_df['match_situation'] = chase_df.apply(classify_match_situation, axis=1)

chase_df[['match_id', 'over', 'ball', 'required_run_rate', 'wickets_left', 'balls_remaining', 'match_situation']].head(30)
```

```python
chase_df['weighted_batter_runs'] = chase_df['batsman_runs'] * chase_df['batter_pressure_score']
chase_df['weighted_batter_boundaries'] = chase_df['is_boundary'] * chase_df['batter_pressure_score']
chase_df['weighted_batter_dismissal'] = chase_df['is_wicket'] * chase_df['batter_pressure_score']
```

```python
batter_stats = chase_df.groupby('batter', as_index=False).agg(
    pressure_balls=('batter_pressure_score', 'count'),
    pressure_exposure=('batter_pressure_score', 'sum'),
    total_runs=('batsman_runs', 'sum'),
    weighted_runs=('weighted_batter_runs', 'sum'),
    total_boundaries=('is_boundary', 'sum'),
    weighted_boundaries=('weighted_batter_boundaries', 'sum'),
    dismissals=('is_wicket', 'sum'),
    weighted_dismissals=('weighted_batter_dismissal', 'sum')
)

batter_stats.head()
```

```python
batter_stats['pressure_avg_score_per_ball'] = (
    batter_stats['pressure_exposure'] / batter_stats['pressure_balls']
)

batter_stats['pressure_adjusted_sr_raw'] = np.where(
    batter_stats['pressure_exposure'] > 0,
    (batter_stats['weighted_runs'] / batter_stats['pressure_exposure']) * 100,
    0
)

batter_stats['pressure_boundary_rate_raw'] = np.where(
    batter_stats['pressure_exposure'] > 0,
    batter_stats['weighted_boundaries'] / batter_stats['pressure_exposure'],
    0
)

batter_stats['pressure_dismissal_rate_raw'] = np.where(
    batter_stats['pressure_exposure'] > 0,
    batter_stats['weighted_dismissals'] / batter_stats['pressure_exposure'],
    0
)

batter_stats[['batter', 'pressure_balls', 'pressure_exposure', 'pressure_adjusted_sr_raw']].head()
```

```python
batter_stats['reliability_raw'] = batter_stats['pressure_balls']
batter_stats[['batter', 'pressure_balls', 'reliability_raw']].head()
```

```python
C_batter = 50
global_batter_sr = batter_stats['pressure_adjusted_sr_raw'].mean()

batter_stats['pressure_adjusted_sr_reliable'] = (
    (batter_stats['pressure_adjusted_sr_raw'] * batter_stats['pressure_balls']) +
    (global_batter_sr * C_batter)
) / (batter_stats['pressure_balls'] + C_batter)

batter_stats[['batter', 'pressure_adjusted_sr_raw', 'pressure_balls', 'pressure_adjusted_sr_reliable']].head(10)
```

```python
batter_stats = batter_stats[batter_stats['pressure_balls'] >= 30].copy()
print("Number of batters after minimum sample filter:", batter_stats.shape[0])
```

```python
from sklearn.preprocessing import MinMaxScaler

batter_norm_cols = [
    'pressure_adjusted_sr_reliable',
    'pressure_exposure',
    'pressure_boundary_rate_raw',
    'reliability_raw'
]

batter_scaler = MinMaxScaler()
batter_stats[[col + '_norm' for col in batter_norm_cols]] = batter_scaler.fit_transform(
    batter_stats[batter_norm_cols]
)

batter_stats.head()
```

```python
dismissal_scaler = MinMaxScaler()
batter_stats['pressure_dismissal_rate_norm_temp'] = dismissal_scaler.fit_transform(
    batter_stats[['pressure_dismissal_rate_raw']]
)

batter_stats['pressure_dismissal_rate_norm'] = 1 - batter_stats['pressure_dismissal_rate_norm_temp']
```

```python
batter_stats['batter_clutch_index'] = (
    0.40 * batter_stats['pressure_adjusted_sr_reliable_norm'] +
    0.20 * batter_stats['pressure_exposure_norm'] +
    0.15 * batter_stats['pressure_boundary_rate_raw_norm'] +
    0.15 * batter_stats['reliability_raw_norm'] +
    0.10 * batter_stats['pressure_dismissal_rate_norm']
)
```

```python
bowler_stats['pressure_avg_score_per_ball'] = (
    bowler_stats['pressure_exposure'] / bowler_stats['pressure_balls']
)

bowler_stats['pressure_economy_raw'] = np.where(
    bowler_stats['pressure_exposure'] > 0,
    (bowler_stats['weighted_runs_conceded'] / bowler_stats['pressure_exposure']) * 6,
    0
)

bowler_stats['pressure_dot_rate_raw'] = np.where(
    bowler_stats['pressure_exposure'] > 0,
    bowler_stats['weighted_dot_balls'] / bowler_stats['pressure_exposure'],
    0
)

bowler_stats['pressure_wicket_rate_raw'] = np.where(
    bowler_stats['pressure_exposure'] > 0,
    bowler_stats['weighted_wickets'] / bowler_stats['pressure_exposure'],
    0
)

bowler_stats[['bowler', 'pressure_balls', 'pressure_exposure', 'pressure_economy_raw']].head()
```

```python
def batter_situation_fit(row, situation):
    if situation == "Death Finish":
        return (
            0.45 * row['death_specialist_score'] +
            0.25 * row['batter_clutch_index'] +
            0.15 * row['accelerator_score'] +
            0.15 * row['reliability_raw_norm']
        )
    elif situation == "High Pressure Chase":
        return (
            0.40 * row['high_rrr_specialist_score'] +
            0.25 * row['batter_clutch_index'] +
            0.20 * row['accelerator_score'] +
            0.15 * row['reliability_raw_norm']
        )
    elif situation == "Collapse Situation":
        return (
            0.40 * row['collapse_specialist_score'] +
            0.25 * row['stabilizer_score'] +
            0.20 * row['batter_clutch_index'] +
            0.15 * row['reliability_raw_norm']
        )
    elif situation == "Controlled Chase":
        return (
            0.35 * row['stabilizer_score'] +
            0.25 * row['batter_clutch_index'] +
            0.20 * row['collapse_specialist_score'] +
            0.20 * row['reliability_raw_norm']
        )
    elif situation == "Comfortable Chase":
        return (
            0.35 * row['batter_clutch_index'] +
            0.25 * row['accelerator_score'] +
            0.20 * row['death_specialist_score'] +
            0.20 * row['reliability_raw_norm']
        )
    else:
        return (
            0.40 * row['batter_clutch_index'] +
            0.20 * row['death_specialist_score'] +
            0.20 * row['high_rrr_specialist_score'] +
            0.20 * row['reliability_raw_norm']
        )
```

```python
def bowler_situation_fit(row, situation):
    if situation == "Death Finish":
        return (
            0.45 * row['death_bowler_score'] +
            0.25 * row['bowler_clutch_index'] +
            0.15 * row['momentum_breaker_score'] +
            0.15 * row['reliability_raw_norm']
        )
    elif situation == "High Pressure Chase":
        return (
            0.35 * row['death_bowler_score'] +
            0.25 * row['momentum_breaker_score'] +
            0.25 * row['bowler_clutch_index'] +
            0.15 * row['reliability_raw_norm']
        )
    elif situation == "Collapse Situation":
        return (
            0.35 * row['set_batter_control_score'] +
            0.25 * row['bowler_clutch_index'] +
            0.20 * row['momentum_breaker_score'] +
            0.20 * row['reliability_raw_norm']
        )
    elif situation == "Controlled Chase":
        return (
            0.35 * row['bowler_clutch_index'] +
            0.30 * row['set_batter_control_score'] +
            0.20 * row['momentum_breaker_score'] +
            0.15 * row['reliability_raw_norm']
        )
    elif situation == "Comfortable Chase":
        return (
            0.40 * row['bowler_clutch_index'] +
            0.25 * row['set_batter_control_score'] +
            0.20 * row['momentum_breaker_score'] +
            0.15 * row['reliability_raw_norm']
        )
    else:
        return (
            0.40 * row['bowler_clutch_index'] +
            0.25 * row['death_bowler_score'] +
            0.20 * row['set_batter_control_score'] +
            0.15 * row['reliability_raw_norm']
        )
```

```python
def batter_role_fit_score(player_role, situation):
    preferred_roles = batter_role_preference.get(situation, [])
    
    if player_role in preferred_roles:
        if preferred_roles.index(player_role) == 0:
            return 1.0
        elif preferred_roles.index(player_role) == 1:
            return 0.8
        else:
            return 0.6
    return 0.3
```

```python
def bowler_role_fit_score(player_role, situation):
    preferred_roles = bowler_role_preference.get(situation, [])
    
    if player_role in preferred_roles:
        if preferred_roles.index(player_role) == 0:
            return 1.0
        elif preferred_roles.index(player_role) == 1:
            return 0.8
        else:
            return 0.6
    return 0.3
```

```python
def recommend_batters_advanced(situation, current_bowler=None, top_n=5):
    col_name = situation.lower().replace(" ", "_") + "_fit"
    
    temp = batter_stats.copy()
    temp['role_fit'] = temp['batter_role'].apply(lambda x: batter_role_fit_score(x, situation))
    temp['base_score'] = (
        0.75 * temp[col_name] +
        0.25 * temp['role_fit']
    )
    
    if current_bowler is not None:
        bowler_matchups = matchup_df[matchup_df['bowler'] == current_bowler][['batter', 'batter_matchup_score']]
        temp = temp.merge(bowler_matchups, on='batter', how='left')
        temp['batter_matchup_score'] = temp['batter_matchup_score'].fillna(0.5)
        temp['final_score'] = 0.80 * temp['base_score'] + 0.20 * temp['batter_matchup_score']
    else:
        temp['final_score'] = temp['base_score']
    
    return temp[[
        'batter',
        'final_score',
        'batter_role',
        'role_fit',
        'risk_reward_profile',
        'reliability_label',
        'strength_tags',
        'weakness_tags'
    ]].sort_values(by='final_score', ascending=False).head(top_n)
```

```python
def suggest_batting_type(situation, wickets_left, required_run_rate):
    if situation == "Collapse Situation":
        return "Send a stabilizer or anchor-type batter."
    elif situation == "Death Finish" and required_run_rate >= 10:
        return "Send a finisher or aggressive accelerator."
    elif situation == "High Pressure Chase":
        return "Send a high-tempo batter who can score quickly."
    elif situation == "Controlled Chase":
        return "Send a balanced or stabilizing batter."
    elif situation == "Comfortable Chase":
        return "Send a batter who can keep momentum without high risk."
    else:
        if wickets_left <= 3:
            return "Prefer a stable batter."
        elif required_run_rate > 10:
            return "Prefer an attacking batter."
        else:
            return "Prefer a balanced batter."
```

```python
def calculate_remaining_overs(bowler_overs, max_overs=4):
    remaining_overs = {}
    
    for bowler, overs in bowler_overs.items():
        remaining = max_overs - overs
        remaining_overs[bowler] = max(0, remaining)
    
    return remaining_overs
```

```python
def get_available_bowlers_with_overs(playing_xi, bowler_overs):
    remaining_overs = calculate_remaining_overs(bowler_overs)
    
    available_bowlers = [
        bowler for bowler in playing_xi
        if remaining_overs.get(bowler, 0) > 0
    ]
    
    return available_bowlers, remaining_overs
```

```python
def filter_bowlers_by_situation(available_bowlers, remaining_overs, situation):
    
    if situation == "Death Finish":
        # Prefer bowlers with at least 2 overs remaining
        filtered = [
            b for b in available_bowlers
            if remaining_overs.get(b, 0) >= 1
        ]
        
        if len(filtered) > 0:
            return filtered
    
    return available_bowlers
```

```python
def recommend_next_bowler_realistic(
    playing_xi,
    bowler_overs,
    situation,
    wickets_left,
    required_run_rate,
    current_batter=None,
    top_n=5
):
    # Step 1: Get valid bowlers
    available_bowlers, remaining_overs = get_available_bowlers_with_overs(
        playing_xi, bowler_overs
    )
    
    if len(available_bowlers) == 0:
        print("No bowlers available (all completed quota)")
        return pd.DataFrame()
    
    # Step 2: Apply situation-based filtering
    filtered_bowlers = filter_bowlers_by_situation(
        available_bowlers, remaining_overs, situation
    )
    
    temp = bowler_stats[bowler_stats['bowler'].isin(filtered_bowlers)].copy()
    
    if temp.empty:
        print("No matching bowlers found in stats")
        return pd.DataFrame()
    
    col_name = situation.lower().replace(" ", "_") + "_fit"
    
    # Step 3: Role fit
    temp['role_fit'] = temp['bowler_role'].apply(
        lambda x: bowler_role_fit_score(x, situation)
    )
    
    temp['base_score'] = 0.75 * temp[col_name] + 0.25 * temp['role_fit']
    
    # Step 4: Matchup boost
    if current_batter is not None:
        batter_matchups = matchup_df[
            matchup_df['batter'] == current_batter
        ][['bowler', 'bowler_matchup_score']]
        
        temp = temp.merge(batter_matchups, on='bowler', how='left')
        temp['bowler_matchup_score'] = temp['bowler_matchup_score'].fillna(0.5)
        
        temp['final_score'] = (
            0.80 * temp['base_score'] +
            0.20 * temp['bowler_matchup_score']
        )
    else:
        temp['final_score'] = temp['base_score']
    
    # Step 5: Add remaining overs info
    temp['remaining_overs'] = temp['bowler'].map(remaining_overs)
    
    return temp.sort_values(by='final_score', ascending=False)[[
        'bowler',
        'final_score',
        'remaining_overs',
        'bowler_role',
        'risk_reward_profile',
        'strength_tags',
        'weakness_tags'
    ]].head(top_n)
```

```python
def recommend_batters_matchup_aware(
    playing_xi,
    situation,
    wickets_left,
    required_run_rate,
    dismissed_players=None,
    current_batters=None,
    current_bowler=None,
    top_n=5
):
    if dismissed_players is None:
        dismissed_players = []
    if current_batters is None:
        current_batters = []
    
    available_batters = get_available_batters(
        playing_xi=playing_xi,
        dismissed_players=dismissed_players,
        current_batters=current_batters
    )
    
    temp = batter_stats[batter_stats['batter'].isin(available_batters)].copy()
    
    if temp.empty:
        print("No matching available batters found in batter_stats.")
        return pd.DataFrame()
    
    col_name = situation.lower().replace(" ", "_") + "_fit"
    
    temp['role_fit'] = temp['batter_role'].apply(lambda x: batter_role_fit_score(x, situation))
    temp['base_score'] = 0.70 * temp[col_name] + 0.20 * temp['role_fit'] + 0.10 * temp['reliability_raw_norm']
    
    if current_bowler is not None:
        bowler_type = get_bowler_type(current_bowler)
        
        direct_scores = []
        direct_confidences = []
        type_scores = []
        
        for batter in temp['batter']:
            d_score, d_conf = get_direct_batter_matchup_score(batter, current_bowler)
            t_score = get_batter_vs_type_score(batter, bowler_type)
            
            direct_scores.append(d_score)
            direct_confidences.append(d_conf)
            type_scores.append(t_score)
        
        temp['direct_matchup_score'] = direct_scores
        temp['direct_matchup_confidence'] = direct_confidences
        temp['type_matchup_score'] = type_scores
        
        temp['combined_matchup_score'] = (
            temp['direct_matchup_score'] * temp['direct_matchup_confidence'] +
            temp['type_matchup_score'] * (1 - temp['direct_matchup_confidence'])
        )
        
        temp['final_score'] = 0.75 * temp['base_score'] + 0.25 * temp['combined_matchup_score']
    else:
        temp['direct_matchup_score'] = 0.5
        temp['direct_matchup_confidence'] = 0.0
        temp['type_matchup_score'] = 0.5
        temp['combined_matchup_score'] = 0.5
        temp['final_score'] = temp['base_score']
    
    return temp[[
        'batter',
        'final_score',
        'batter_role',
        'risk_reward_profile',
        'reliability_label',
        'strength_tags',
        'weakness_tags',
        'direct_matchup_score',
        'direct_matchup_confidence',
        'type_matchup_score',
        'combined_matchup_score'
    ]].sort_values(by='final_score', ascending=False).head(top_n)
```

```python
def recommend_bowlers_matchup_aware(
    playing_xi,
    bowler_overs,
    situation,
    wickets_left,
    required_run_rate,
    current_batter=None,
    top_n=5
):
    available_bowlers, remaining_overs = get_available_bowlers_with_overs(playing_xi, bowler_overs)
    
    if len(available_bowlers) == 0:
        print("No bowlers available.")
        return pd.DataFrame()
    
    filtered_bowlers = filter_bowlers_by_situation(available_bowlers, remaining_overs, situation)
    
    temp = bowler_stats[bowler_stats['bowler'].isin(filtered_bowlers)].copy()
    
    if temp.empty:
        print("No matching bowlers found in bowler_stats.")
        return pd.DataFrame()
    
    col_name = situation.lower().replace(" ", "_") + "_fit"
    
    temp['role_fit'] = temp['bowler_role'].apply(lambda x: bowler_role_fit_score(x, situation))
    temp['base_score'] = 0.70 * temp[col_name] + 0.20 * temp['role_fit'] + 0.10 * temp['reliability_raw_norm']
    
    if current_batter is not None:
        batter_hand = get_batter_hand(current_batter)
        
        direct_scores = []
        direct_confidences = []
        hand_scores = []
        
        for bowler in temp['bowler']:
            row = matchup_df[(matchup_df['batter'] == current_batter) & (matchup_df['bowler'] == bowler)]
            
            if row.empty:
                d_score = 0.5
                d_conf = 0.0
            else:
                d_score = row.iloc[0]['bowler_matchup_score']
                d_conf = row.iloc[0]['matchup_confidence_score']
            
            h_score = get_bowler_vs_hand_score(bowler, batter_hand)
            
            direct_scores.append(d_score)
            direct_confidences.append(d_conf)
            hand_scores.append(h_score)
        
        temp['direct_matchup_score'] = direct_scores
        temp['direct_matchup_confidence'] = direct_confidences
        temp['hand_matchup_score'] = hand_scores
        
        temp['combined_matchup_score'] = (
            temp['direct_matchup_score'] * temp['direct_matchup_confidence'] +
            temp['hand_matchup_score'] * (1 - temp['direct_matchup_confidence'])
        )
        
        temp['final_score'] = 0.75 * temp['base_score'] + 0.25 * temp['combined_matchup_score']
    else:
        temp['direct_matchup_score'] = 0.5
        temp['direct_matchup_confidence'] = 0.0
        temp['hand_matchup_score'] = 0.5
        temp['combined_matchup_score'] = 0.5
        temp['final_score'] = temp['base_score']
    
    temp['remaining_overs'] = temp['bowler'].map(remaining_overs)
    
    return temp[[
        'bowler',
        'final_score',
        'remaining_overs',
        'bowler_role',
        'risk_reward_profile',
        'reliability_label',
        'strength_tags',
        'weakness_tags',
        'direct_matchup_score',
        'direct_matchup_confidence',
        'hand_matchup_score',
        'combined_matchup_score'
    ]].sort_values(by='final_score', ascending=False).head(top_n)
```

```python
def filter_bowlers_by_advanced_legality(
    available_bowlers,
    remaining_overs,
    situation,
    current_over,
    previous_over_bowler=None,
    save_death_over=False,
    death_bowlers=None
):
    if death_bowlers is None:
        death_bowlers = []

    filtered = available_bowlers.copy()

    # Remove bowlers with no overs left
    filtered = [b for b in filtered if remaining_overs.get(b, 0) > 0]

    # Optional: if simulating one-bowler-per-over sequence, avoid same bowler if user wants that logic
    if previous_over_bowler is not None:
        filtered = [b for b in filtered if b != previous_over_bowler]

    # Save one over of key death bowler for over 20
    if save_death_over and current_over < 20:
        filtered = [
            b for b in filtered
            if not (b in death_bowlers and remaining_overs.get(b, 0) == 1 and current_over < 19)
        ] or filtered

    # In death overs, prefer bowlers with death skill
    if situation == "Death Finish":
        death_pref = [b for b in filtered if b in death_bowlers]
        if len(death_pref) > 0:
            return death_pref

    return filtered
```

```python
def recommend_bowlers_matchup_aware_advanced(
    playing_xi,
    bowler_overs,
    situation,
    wickets_left,
    required_run_rate,
    current_over,
    current_batter=None,
    previous_over_bowler=None,
    save_death_over=False,
    death_bowlers=None,
    top_n=5
):
    available_bowlers, remaining_overs = get_available_bowlers_with_overs(playing_xi, bowler_overs)

    if len(available_bowlers) == 0:
        print("No bowlers available.")
        return pd.DataFrame()

    filtered_bowlers = filter_bowlers_by_advanced_legality(
        available_bowlers=available_bowlers,
        remaining_overs=remaining_overs,
        situation=situation,
        current_over=current_over,
        previous_over_bowler=previous_over_bowler,
        save_death_over=save_death_over,
        death_bowlers=death_bowlers
    )

    temp = bowler_stats[bowler_stats['bowler'].isin(filtered_bowlers)].copy()

    if temp.empty:
        print("No matching bowlers found in bowler_stats.")
        return pd.DataFrame()

    col_name = situation.lower().replace(" ", "_") + "_fit"

    temp['role_fit'] = temp['bowler_role'].apply(lambda x: bowler_role_fit_score(x, situation))
    temp['base_score'] = 0.70 * temp[col_name] + 0.20 * temp['role_fit'] + 0.10 * temp['reliability_raw_norm']

    if current_batter is not None:
        batter_hand = get_batter_hand(current_batter)

        direct_scores = []
        direct_confidences = []
        hand_scores = []

        for bowler in temp['bowler']:
            row = matchup_df[(matchup_df['batter'] == current_batter) & (matchup_df['bowler'] == bowler)]

            if row.empty:
                d_score = 0.5
                d_conf = 0.0
            else:
                d_score = row.iloc[0]['bowler_matchup_score']
                d_conf = row.iloc[0]['matchup_confidence_score']

            h_score = get_bowler_vs_hand_score(bowler, batter_hand)

            direct_scores.append(d_score)
            direct_confidences.append(d_conf)
            hand_scores.append(h_score)

        temp['direct_matchup_score'] = direct_scores
        temp['direct_matchup_confidence'] = direct_confidences
        temp['hand_matchup_score'] = hand_scores

        temp['combined_matchup_score'] = (
            temp['direct_matchup_score'] * temp['direct_matchup_confidence'] +
            temp['hand_matchup_score'] * (1 - temp['direct_matchup_confidence'])
        )

        temp['final_score'] = 0.75 * temp['base_score'] + 0.25 * temp['combined_matchup_score']
    else:
        temp['direct_matchup_score'] = 0.5
        temp['direct_matchup_confidence'] = 0.0
        temp['hand_matchup_score'] = 0.5
        temp['combined_matchup_score'] = 0.5
        temp['final_score'] = temp['base_score']

    temp['remaining_overs'] = temp['bowler'].map(remaining_overs)

    return temp[[
        'bowler',
        'final_score',
        'remaining_overs',
        'bowler_role',
        'risk_reward_profile',
        'reliability_label',
        'strength_tags',
        'weakness_tags',
        'direct_matchup_score',
        'direct_matchup_confidence',
        'hand_matchup_score',
        'combined_matchup_score'
    ]].sort_values(by='final_score', ascending=False).head(top_n)
```

```python
def next_bowler_assistant_matchup_aware_advanced(
    playing_xi,
    bowler_overs,
    situation,
    wickets_left,
    required_run_rate,
    current_over,
    current_batter=None,
    previous_over_bowler=None,
    save_death_over=False,
    death_bowlers=None,
    top_n=5
):
    print("=" * 90)
    print("ADVANCED MATCHUP-AWARE NEXT BOWLER ASSISTANT")
    print("=" * 90)
    print(f"Situation: {situation}")
    print(f"Current Over: {current_over}")
    print(f"Wickets Left: {wickets_left}")
    print(f"Required RR: {required_run_rate}")
    print(f"Current Batter: {current_batter}")
    print(f"Previous Over Bowler: {previous_over_bowler}")
    print(f"Save Death Bowler Over: {save_death_over}")
    print()

    recs = recommend_bowlers_matchup_aware_advanced(
        playing_xi=playing_xi,
        bowler_overs=bowler_overs,
        situation=situation,
        wickets_left=wickets_left,
        required_run_rate=required_run_rate,
        current_over=current_over,
        current_batter=current_batter,
        previous_over_bowler=previous_over_bowler,
        save_death_over=save_death_over,
        death_bowlers=death_bowlers,
        top_n=top_n
    )

    display(recs)
    return recs
```

```python
def recommend_batters_matchup_aware(
    playing_xi,
    situation,
    wickets_left,
    required_run_rate,
    dismissed_players=None,
    current_batters=None,
    current_bowler=None,
    top_n=5
):
    if dismissed_players is None:
        dismissed_players = []
    if current_batters is None:
        current_batters = []

    available_batters = get_available_batters(
        playing_xi=playing_xi,
        dismissed_players=dismissed_players,
        current_batters=current_batters
    )

    temp = batter_stats[batter_stats['batter'].isin(available_batters)].copy()

    if temp.empty:
        print("No matching available batters found in batter_stats.")
        return pd.DataFrame()

    col_name = situation.lower().replace(" ", "_") + "_fit"

    temp['role_fit'] = temp['batter_role'].apply(lambda x: batter_role_fit_score(x, situation))
    temp['batting_priority'] = temp['batter'].apply(get_batting_priority)

    # If at least one real batting option exists, drop extreme tailenders
    if (temp['batting_priority'] >= 0.60).any():
        temp = temp[temp['batting_priority'] >= 0.30].copy()

    weights = get_situation_weights(situation)

    temp['base_score'] = (
        weights["situation"] * temp[col_name] +
        weights["role"] * temp['role_fit'] +
        weights["reliability"] * temp['reliability_raw_norm'] +
        weights["batting_priority"] * temp['batting_priority']
    )

    if current_bowler is not None:
        bowler_type = get_bowler_type(current_bowler)

        direct_scores = []
        direct_confidences = []
        type_scores = []

        for batter in temp['batter']:
            d_score, d_conf = get_direct_batter_matchup_score(batter, current_bowler)
            t_score = get_batter_vs_type_score(batter, bowler_type)

            direct_scores.append(d_score)
            direct_confidences.append(d_conf)
            type_scores.append(t_score)

        temp['direct_matchup_score'] = direct_scores
        temp['direct_matchup_confidence'] = direct_confidences
        temp['type_matchup_score'] = type_scores

        temp['combined_matchup_score'] = (
            temp['direct_matchup_score'] * temp['direct_matchup_confidence'] +
            temp['type_matchup_score'] * (1 - temp['direct_matchup_confidence'])
        )

        temp['final_score'] = (
            0.80 * temp['base_score'] +
            0.20 * temp['combined_matchup_score']
        )
    else:
        temp['direct_matchup_score'] = 0.5
        temp['direct_matchup_confidence'] = 0.0
        temp['type_matchup_score'] = 0.5
        temp['combined_matchup_score'] = 0.5
        temp['final_score'] = temp['base_score']

    return temp[[
        'batter',
        'final_score',
        'batting_priority',
        'batter_role',
        'risk_reward_profile',
        'reliability_label',
        'strength_tags',
        'weakness_tags',
        'direct_matchup_score',
        'direct_matchup_confidence',
        'type_matchup_score',
        'combined_matchup_score'
    ]].sort_values(by='final_score', ascending=False).head(top_n)
```

```python
# STEP 1: Imports

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# STEP 2: make a copy and create target label

clf_df = batter_stats.copy()

clf_df['suitable_batter'] = np.where(
    (clf_df['batter_clutch_index'] >= clf_df['batter_clutch_index'].median()) &
    (clf_df['pressure_adjusted_sr_reliable'] >= clf_df['pressure_adjusted_sr_reliable'].median()) &
    (clf_df['pressure_dismissal_rate_raw'] <= clf_df['pressure_dismissal_rate_raw'].median()),
    1, 0
)

clf_df[['batter', 'batter_clutch_index', 'pressure_adjusted_sr_reliable',
        'pressure_dismissal_rate_raw', 'suitable_batter']].head(10)
```

```python
clf_df['suitable_batter'].value_counts()
```

```python
# STEP 3: select input features

feature_cols = [
    'pressure_adjusted_sr_reliable',
    'pressure_exposure',
    'pressure_boundary_rate_raw',
    'pressure_dismissal_rate_raw',
    'batter_clutch_index',
    'death_specialist_score',
    'high_rrr_specialist_score',
    'collapse_specialist_score',
    'stabilizer_score',
    'accelerator_score',
    'pp_strike_rate',
    'middle_strike_rate',
    'death_strike_rate'
]

X = clf_df[feature_cols]
y = clf_df['suitable_batter']

print("X shape:", X.shape)
print("y shape:", y.shape)
print(X.head())
```

```python
# STEP 4: split data into train and test

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,      # 80% train, 20% test
    random_state=42,
    stratify=y          # keeps 0/1 balance same
)

print("Training data:", X_train.shape)
print("Testing data:", X_test.shape)
```

```python
# STEP 5: define classifiers

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    
    "Decision Tree": DecisionTreeClassifier(max_depth=5),
    
    "Random Forest": RandomForestClassifier(
        n_estimators=50,   # kept small so it runs fast
        max_depth=5,
        random_state=42
    ),
    
    "Gradient Boosting": GradientBoostingClassifier(
        n_estimators=50,   # kept small for speed
        learning_rate=0.1,
        random_state=42
    )
}

print("Models ready:", list(models.keys()))
```

```python
# STEP 6: train all models and evaluate

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    results[name] = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred, zero_division=0),
        'Recall': recall_score(y_test, y_pred, zero_division=0),
        'F1 Score': f1_score(y_test, y_pred, zero_division=0)
    }

results_df = pd.DataFrame(results).T
results_df = results_df.sort_values(by='F1 Score', ascending=False)

print(results_df)
```

```python
# BOWLER CLASSIFIER - STEP 3
# select features

bowler_feature_cols = [
    'pressure_wicket_rate_raw',
    'pressure_economy_raw',
    'pressure_dot_rate_raw',
    'bowler_clutch_index',
    'death_bowler_score',
    'pp_economy',
    'middle_economy',
    'death_economy'
]

X_bowler = bowler_clf_df[bowler_feature_cols]
y_bowler = bowler_clf_df['suitable_bowler']

print("X shape:", X_bowler.shape)
print("y shape:", y_bowler.shape)
print(X_bowler.head())
```

```python
# BOWLER CLASSIFIER - STEP 4
# fill missing values, then split train/test

X_bowler = X_bowler.fillna(X_bowler.median(numeric_only=True))

print("Missing values after fill:")
print(X_bowler.isna().sum())

Xb_train, Xb_test, yb_train, yb_test = train_test_split(
    X_bowler, y_bowler,
    test_size=0.2,
    random_state=42,
    stratify=y_bowler
)

print("Training data:", Xb_train.shape)
print("Testing data:", Xb_test.shape)
```

```python
# BOWLER CLASSIFIER - STEP 5
# define models

bowler_models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    
    "Decision Tree": DecisionTreeClassifier(max_depth=5),
    
    "Random Forest": RandomForestClassifier(
        n_estimators=50,
        max_depth=5,
        random_state=42
    ),
    
    "Gradient Boosting": GradientBoostingClassifier(
        n_estimators=50,
        learning_rate=0.1,
        random_state=42
    )
}

print("Bowler models ready:", list(bowler_models.keys()))
```

```python
# BOWLER CLASSIFIER - STEP 6
# train all bowler models and evaluate

bowler_results = {}

for name, model in bowler_models.items():
    model.fit(Xb_train, yb_train)
    yb_pred = model.predict(Xb_test)

    bowler_results[name] = {
        'Accuracy': accuracy_score(yb_test, yb_pred),
        'Precision': precision_score(yb_test, yb_pred, zero_division=0),
        'Recall': recall_score(yb_test, yb_pred, zero_division=0),
        'F1 Score': f1_score(yb_test, yb_pred, zero_division=0)
    }

bowler_results_df = pd.DataFrame(bowler_results).T
bowler_results_df = bowler_results_df.sort_values(by='F1 Score', ascending=False)

print(bowler_results_df)
```

```python
# BOWLER CLASSIFIER - STEP 7

best_bowler_model_name = bowler_results_df.index[0]
best_bowler_model = bowler_models[best_bowler_model_name]

best_bowler_model.fit(Xb_train, yb_train)

print("Best bowler model:", best_bowler_model_name)
```

```python
# INTEGRATION STEP 1
# add prediction column to batter_stats

batter_stats['classifier_suitable'] = best_model.predict(
    batter_stats[feature_cols].fillna(batter_stats[feature_cols].median())
)

batter_stats[['batter', 'classifier_suitable']].head()
```

```python
# INTEGRATION STEP 2
# convert classifier output into a score/bonus

batter_stats['classifier_score'] = batter_stats['classifier_suitable'].apply(
    lambda x: 1.0 if x == 1 else 0.0
)

batter_stats[['batter', 'classifier_suitable', 'classifier_score']].head(10)
```

```python
# INTEGRATION STEP 3
# combine existing performance score + classifier score

batter_stats['final_batter_reco_score'] = (
    0.50 * batter_stats['batter_clutch_index'] +
    0.20 * batter_stats['classifier_score'] +
    0.15 * batter_stats['accelerator_score'] +
    0.15 * batter_stats['stabilizer_score']
)

batter_stats[['batter', 'batter_clutch_index', 'classifier_score',
              'accelerator_score', 'stabilizer_score',
              'final_batter_reco_score']].sort_values(
    by='final_batter_reco_score', ascending=False
).head(10)
```

```python
# INTEGRATION STEP 4
# simple batter recommendation function using final score

def recommend_batters_with_classifier(top_n=5):
    cols_to_show = [
        'batter',
        'final_batter_reco_score',
        'batter_clutch_index',
        'classifier_score',
        'accelerator_score',
        'stabilizer_score'
    ]
    
    temp = batter_stats.sort_values(
        by='final_batter_reco_score',
        ascending=False
    ).copy()
    
    return temp[cols_to_show].head(top_n)

recommend_batters_with_classifier(5)
```

```python
# INTEGRATION STEP 5
# add classifier prediction to bowler_stats

bowler_stats['classifier_suitable'] = best_bowler_model.predict(
    bowler_stats[bowler_feature_cols].fillna(
        bowler_stats[bowler_feature_cols].median()
    )
)

bowler_stats[['bowler', 'classifier_suitable']].head(10)
```

```python
# INTEGRATION STEP 6
# convert bowler classifier output to score

bowler_stats['bowler_classifier_score'] = bowler_stats['classifier_suitable'].apply(
    lambda x: 1.0 if x == 1 else 0.0
)

bowler_stats[['bowler', 'classifier_suitable', 'bowler_classifier_score']].head(10)
```

```python
# INTEGRATION STEP 7
# combine bowler cricket metrics + classifier score

bowler_stats['final_bowler_reco_score'] = (
    0.50 * bowler_stats['bowler_clutch_index'] +
    0.20 * bowler_stats['bowler_classifier_score'] +
    0.20 * bowler_stats['death_bowler_score'] +
    0.10 * bowler_stats['pressure_dot_rate_raw']
)

bowler_stats[['bowler', 'bowler_clutch_index', 'bowler_classifier_score',
              'death_bowler_score', 'pressure_dot_rate_raw',
              'final_bowler_reco_score']].sort_values(
    by='final_bowler_reco_score', ascending=False
).head(10)
```

```python
# INTEGRATION STEP 8
# simple bowler recommendation function using final score

def recommend_bowlers_with_classifier(top_n=5):
    cols_to_show = [
        'bowler',
        'final_bowler_reco_score',
        'bowler_clutch_index',
        'bowler_classifier_score',
        'death_bowler_score',
        'pressure_dot_rate_raw'
    ]
    
    temp = bowler_stats.sort_values(
        by='final_bowler_reco_score',
        ascending=False
    ).copy()
    
    return temp[cols_to_show].head(top_n)

recommend_bowlers_with_classifier(5)
```
